I broke the problem down into a main function, a menu function, a choice function for players and a choice function for the computer. The menu function prints out the choices for the player and waits for a response. I then use my choice function for players to continually prompt the user for a valid choice and do not return until I have one. The choice function for the computer takes in a dictionary with the keys being "rock, paper or scissors" and the associated values with those keys are counters for how many times the user picks each one. The computer picks the most used choice by the user and then returns the choice that will beat it. The main function has a while loop that runs until the user enters quit. Inside the loop I do all the checking to see which player won. I make calls to the other functions within the main while loop to reuse code.

```python
import random
def run_game():
    """The main function that runs the game"""
    player_choice = menu()
    computer_choice = random.randint(1,3) #Random choice for first turn
    player_wins = 0
    rounds = 0
    totals = {'rock': 0, 'paper': 0, 'scissors': 0}
    while player_choice != 'quit': #choice 4 is for quit
        rounds += 1
        #Compare the player choice to computer's choice
        if player_choice == 'rock': #Player chose rock
            totals['rock'] += 1
            if computer_choice == 'rock': #Draw
                print "We both chose rock. Draw."
            elif computer_choice == 'scissors': #Rock beats scissors
                player_wins += 1
                print "You won this round! Rock beats scissors!"
            else: #Paper beats rock
                print "Computer wins this round! Paper beats rock!"
        elif player_choice == 'paper': #Player chose paper
            totals['paper'] += 1
            if computer_choice == 'paper': #Draw
                print "We both chose paper. Draw."
            elif computer_choice == 'rock': #Paper beats rock
                player_wins += 1
                print "You won this round! Paper beats rock!"
            else: #Scissors beats paper
                print "Computer wins this round! Scissors beats paper!"
        elif player_choice == 'scissors': #Player chose scissors 
            totals['scissors'] += 1
            if computer_choice == 'scissors': #Draw
                print "We both chose scissors. Draw."
            elif computer_choice == 'paper': #Scissors beats paper
                player_wins += 1
                print "You won this round! Scissors beats paper!"
            else: #Rock beats scissors
                print "Computer wins this round! Rock beats scissors!"

        computer_choice = get_choice_computer(totals)    
        player_choice = menu()
        
    print "Thank you for playing."
    print "Total rounds played: " + str(rounds)
    print "Total wins by player: " + str(player_wins)
    
def menu():
    """Print the four choices for the player.
    Returns the player choice as a string."""
    print
    print "Your options are:"
    print "1) Rock"
    print "2) Paper"
    print "3) Scissors"
    print "4) Quit"
    print
    return get_choice("Enter your choice: ")

def get_choice(prompt):
    """Prompt the user for a choice using any given prompt.
    Will return once a valid choice has been made."""
    choice = raw_input(prompt).lower()
    while choice not in ('rock', 'paper', 'scissors', 'quit'): #Valid choices
        print "Invalid choice. Please choose again."
        choice = raw_input(prompt).lower()
    return choice

def get_choice_computer(counts):
    """Picks most common human choice using max function.
    Parameter is dictionary with keys being choices and values being counts of each choice.
    Assigns computer choice to beat human's favorite choice.
    Returns computer choice as string"""
    
    common = max(counts, key=counts.get) #Should return key with highest value
    com_choice = 'rock' #Assign rock for base. Might help prevent error
    
    if common == 'rock': #Player keeps choosing rock
        com_choice = 'paper' #Assign paper for computer
    elif common == 'paper': #Player keeps choosing paper
        com_choice = 'scissors' #Assign scissors for computer
    elif common == 'scissors': #Player keeps choosing scissors
        com_choice = 'rock' #Assign rock for computer
    return com_choice
```
